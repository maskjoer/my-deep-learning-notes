```python
import pandas as pd
from sklearn.datasets import load_iris
import mglearn
import numpy as np
iris_dataset = load_iris()
print(iris_dataset.keys())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data,iris_dataset.target,random_state=0)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
iris_dataframe =pd.DataFrame(X_train,columns = iris_dataset.feature_names)
grr =pd.plotting.scatter_matrix(iris_dataframe,c=y_train,figsize=(20,20), marker='0', hist_kwds={'bins':20}, s =60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
X_new =np.array([[5,2.9,1,0.2]])
X_new.shape
```

```python
prediction = knn.predict(X_new)
iris_dataset['target_names'][prediction]
```

```python
y_pred = knn.predict(X_test)
print(y_pred)
```

```python
print("{:.2f}".format(np.mean(y_pred==y_test)))
print("{:.2f}".format(knn.score(X_test,y_test)))
```
